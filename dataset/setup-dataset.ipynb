```python
# Import necessary libraries
import pandas as pd
import os
from google.colab import drive
import glob

# Mount Google Drive with force remount
drive.mount('/content/drive', force_remount=True)

# Specify the root directory containing the folders with CSV files
root_dir = 'drive/MyDrive/Legal_QA_HRAG_Project/dataset/dataset-folder'  # Adjusted for correct "My Drive" spacing

# Debug: Verify if the directory exists
if not os.path.exists(root_dir):
    print(f"Error: Directory {root_dir} does not exist. Please check the path.")
else:
    print(f"Directory {root_dir} exists. Checking contents:")

    # List directory structure
    for folder, subfolders, files in os.walk(root_dir):
        print(f"Folder: {folder}")
        for file in files:
            print(f"  File: {file}")
        for subfolder in subfolders:
            print(f"  Subfolder: {subfolder}")

# Initialize an empty list to store DataFrames
all_dataframes = []

# Use glob to find all CSV files recursively (case-insensitive)
csv_files = glob.glob(os.path.join(root_dir, '**/*.[cC][sS][vV]'), recursive=True)

if not csv_files:
    print("No CSV files found in the specified directory.")
else:
    print(f"Found {len(csv_files)} CSV files:")
    for file_path in csv_files:
        print(f"Processing file: {file_path}")
        try:
            # Try different encodings to handle potential mismatches
            try:
                df = pd.read_csv(file_path, encoding='utf-8')
            except UnicodeDecodeError:
                df = pd.read_csv(file_path, encoding='utf-8-sig')
            all_dataframes.append(df)
        except Exception as e:
            print(f"Error reading {file_path}: {e}")

# Combine all DataFrames into one
if all_dataframes:
    combined_df = pd.concat(all_dataframes, ignore_index=True)
    print(f"Total records combined: {len(combined_df)}")

    # Save the combined DataFrame to a new CSV file
    output_path = 'drive/MyDrive/Legal_QA_HRAG_Project/dataset/copus-legal-vi.csv'
    combined_df.to_csv(output_path, index=False, encoding='utf-8')
    print(f"Combined CSV saved to: {output_path}")
else:
    print("No CSV files were processed.")
```

```python
# Import các thư viện cần thiết
import pandas as pd
import os

# Định nghĩa đường dẫn đến file trên Google Drive
BASE_DIR = '/content/drive/MyDrive/Legal_QA_HRAG_Project/dataset'
ALQAC_FILE = os.path.join(BASE_DIR, 'train_SweetMask_legal_qa_vi.csv')
COPUS_FILE = os.path.join(BASE_DIR, 'copus-legal-vi.csv')
OUTPUT_FILE = os.path.join(BASE_DIR, 'combined_legal_qa.csv')

# Kết nối với Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Đọc hai file CSV
print("Đang đọc file train_SweetMask_legal_qa_vi.csv...")
alqac_df = pd.read_csv(ALQAC_FILE)
print(f"train_SweetMask_legal_qa_vi.csv có {len(alqac_df)} record với các cột: {alqac_df.columns.tolist()}")

print("Đang đọc file copus-legal-vi.csv...")
copus_df = pd.read_csv(COPUS_FILE)
print(f"copus-legal-vi.csv có {len(copus_df)} record với các cột: {copus_df.columns.tolist()}")

# Kiểm tra cột cơ bản (context, question, answer)
required_columns = ['context', 'question', 'answer']
for df in [alqac_df, copus_df]:
    missing_columns = [col for col in required_columns if col not in df.columns]
    if missing_columns:
        print(f"Cảnh báo: Các cột thiếu trong {df}: {missing_columns}. Hãy kiểm tra lại file!")
        raise ValueError("Cấu trúc cột không khớp.")

# Gộp hai dataframe
combined_df = pd.concat([alqac_df, copus_df], ignore_index=True)
print(f"Đã gộp thành công, tổng số record: {len(combined_df)}")

# Xóa các dòng trùng lặp (nếu có)
combined_df = combined_df.drop_duplicates(subset=['context', 'question', 'answer'], keep='first')
print(f"Sau khi xóa trùng lặp, còn {len(combined_df)} record")

# Lưu file mới
combined_df.to_csv(OUTPUT_FILE, index=False, encoding='utf-8-sig')
print(f"Đã lưu file kết quả tại: {OUTPUT_FILE}")

# Kiểm tra nội dung file mới
print("\nNội dung đầu tiên của file mới:")
print(combined_df.head())
```

```python
# Import các thư viện cần thiết
import pandas as pd
import os

# Định nghĩa đường dẫn đến file trên Google Drive
BASE_DIR = '/content/drive/MyDrive/Legal_QA_HRAG_Project/dataset'
INPUT_FILE = os.path.join(BASE_DIR, 'train_SweetMask_legal_qa_vi.parquet')
OUTPUT_FILE = os.path.join(BASE_DIR, 'train_SweetMask_legal_qa_vi.csv')  # Lưu dưới dạng CSV

# Kết nối với Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Đọc file Parquet
try:
    print("Đang đọc file train_SweetMask_legal_qa_vi.parquet...")
    df = pd.read_parquet(INPUT_FILE)
    print(f"Đã đọc thành công. File có {len(df)} record với các cột: {df.columns.tolist()}")
    print("\nNội dung đầu tiên của file:")
    print(df.head())
except FileNotFoundError:
    print(f"Lỗi: Không tìm thấy file {INPUT_FILE}. Vui lòng kiểm tra đường dẫn hoặc tải file lên!")
except Exception as e:
    print(f"Lỗi khi đọc file: {str(e)}. Vui lòng kiểm tra định dạng file!")

# Kiểm tra và xử lý cột
# Giả định file có ít nhất cột question và answer (hoặc tương tự)
required_columns = ['question', 'answer']
missing_columns = [col for col in required_columns if col not in df.columns]
if missing_columns:
    print(f"Cảnh báo: Các cột thiếu: {missing_columns}. Hãy kiểm tra lại cấu trúc file!")
    raise ValueError("Cấu trúc cột không khớp với yêu cầu tối thiểu.")

# Xử lý cột context (vì không có cột context)
# Lựa chọn 1: Ghép question và answer làm context tạm thời
df['context'] = df['answer']

# (Tùy chọn) Nếu bạn có nguồn dữ liệu khác để tạo context, hãy thay thế dòng trên
# Ví dụ: Nếu có cột 'text' hoặc 'passage', dùng nó làm context
# df['context'] = df['text']  # Thay 'text' bằng tên cột thực tế nếu có

# Tạo dataframe mới với 3 cột
new_df = df[['context', 'question', 'answer']].copy()

# Xóa các dòng có giá trị thiếu (nếu có)
new_df = new_df.dropna()
print(f"Sau khi xử lý, còn {len(new_df)} record")

# Lưu file mới dưới dạng CSV
new_df.to_csv(OUTPUT_FILE, index=False, encoding='utf-8-sig')
print(f"Đã lưu file kết quả tại: {OUTPUT_FILE}")

# Kiểm tra nội dung file mới
print("\nNội dung đầu tiên của file mới:")
print(new_df.head())
```
